```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

Fishy = np.random.poisson(5, 100)
plt.hist(Fishy)
plt.axvline(Fishy.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(Fishy.mean() + Fishy.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(Fishy.mean() - Fishy.std(), color='r', linestyle='dashed', linewidth=2)
plt.title('Poisson Distribution')
plt.show

```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

Hulk = np.random.gamma(10, 10, 100)
plt.hist(Hulk)
plt.axvline(Hulk.mean(), color='y', linestyle='solid', linewidth=2)
plt.axvline(Hulk.mean() + Hulk.std(), color='y', linestyle='dashed', linewidth=2)
plt.axvline(Hulk.mean() - Hulk.std(), color='y', linestyle='dashed', linewidth=2)
plt.title('Gamma Distribution')
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

Lucky = np.random.binomial(45, 0.4, 100)
plt.hist(Lucky, color='r')
plt.axvline(Lucky.mean(), color='black', linestyle='solid', linewidth=2)
plt.axvline(Lucky.mean() + Lucky.std(), color='black', linestyle='dashed', linewidth=2)
plt.axvline(Lucky.mean() - Lucky.std(), color='black', linestyle='dashed', linewidth=2)
plt.title('Binomial Distribution')
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

RV1 = np.random.normal(5, 0.5, 100)
RV2 = np.random.normal(10, 1, 100)
RV3 = RV1 + RV2
plt.hist(RV3, color='purple')
plt.axvline(RV3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(RV3.mean() + RV3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(RV3.mean() - RV3.std(), color='b', linestyle='dashed', linewidth=2)
plt.title('Random variable as a sum of two random variables')
plt.show
```
